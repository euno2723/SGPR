```python
import numpy as np
import pandas as pd
import GPy
import pods
```

```python
np.random.seed(seed=1)
```

```python
data = pods.datasets.olympic_100m_men()
X, Y = data["X"], data["Y"]
X_pred = np.linspace(X[:,0].min() - 30,
                     X[:,0].max() + 30,
                     500).reshape(-1,1)
```

```python
pd.concat([pd.DataFrame(X.T, index=["X"]), pd.DataFrame(Y.T, index=["Y"])], axis=0)
```

```python
k = GPy.kern.RBF(input_dim=1,
                 variance=1.,
                 lengthscale=0.1)
k
```

```python
import numpy as np
import pandas as pd
import GPy
import pods

np.random.seed(seed=1)


data = pods.datasets.olympic_100m_men()
X, Y = data["X"], data["Y"]
X_pred = np.linspace(X[:,0].min() - 30,
                     X[:,0].max() + 30,
                     500).reshape(-1,1)

k = GPy.kern.RBF(input_dim=1,
                 variance=1.,
                 lengthscale=10.)

m = GPy.models.GPRegression(X, Y, k)
mu, var = m.predict(X_pred)
pd.concat([pd.DataFrame(mu.T, index=["mu"]), pd.DataFrame(var.T, index=["var"])], axis=0)
```
